```python
from collections import deque

def bfs(start:list, end:list):
    dr = [-1, -1, 1, 1, -2, -2, 2, 2]
    dc = [2, -2, 2, -2, 1, -1, 1, -1]
    q = deque()
    q.append(start)
    matrix[start[0]][start[1]] = 1
    while q:
        r,c = q.popleft()
        for i in range(8):
            r_moved = r + dr[i]
            c_moved = c + dc[i]
            if (
                0 <= r_moved < length
                and 0 <= c_moved < length
            ):
                if matrix[r_moved][c_moved] == 0:
                    matrix[r_moved][c_moved] = 1
                    depth[r_moved][c_moved] = depth[r][c] + 1
                    if [r_moved, c_moved] == end:
                        return depth[r_moved][c_moved]
                    q.append([r_moved, c_moved])                
    return depth[end[0]][end[1]]
rep = int(input())
for _ in range(rep):
    length = int(input())
    depth = [[0]*length for _ in range(length)]
    matrix = [[0]*length for _ in range(length)]
    start = list(map(int, input().split()))
    end = list(map(int, input().split()))
    print(bfs(start,end))


                    
        
    
```
