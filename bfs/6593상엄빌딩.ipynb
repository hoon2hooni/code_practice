```python
from collections import deque
def check_and_go(moved: list,before_: list,q):
    f_,r_,c_ = moved
    f,r,c = before_
    if (
        0 <= r_ < row_count
        and 0 <= c_ < col_count
        and 0 <= f_ < floor_count
    ):
        if building[f_][r_][c_] == '.':
            building[f_][r_][c_]  = '#'
            depth[f_][r_][c_] = depth[f][r][c] + 1
            q.append([f_,r_,c_])
        elif building[f_][r_][c_] == 'E':
            return True
    
    return False



def bfs(start: list):
    q = deque()
    q.append(start)
    updated_f = start[2]
    while q:
        before =  q.popleft()
        f, r ,c = before 
        pprint.pprint(building, width = 50)
        pprint.pprint(depth, width = 50)
        for i in range(4):
            #상하좌우
            updated_r = r + dr[i]
            updated_c = c + dc[i]
            updated =[f,updated_r,updated_c]
            if check_and_go(updated,before,q) is True:
                return f"Escaped in {depth[f][r][c] + 1} minute(s)"
        for i in range(2):
            #층 위아래
            updated_f = f + df[i]
            updated =[updated_f,r,c]
            if check_and_go(updated,before,q) is True:
                return f"Escaped in {depth[f][r][c] + 1} minute(s)"
    return "Trapped!"

floor_count, row_count, col_count = map(int,input().split())
building =[]
for _ in range(floor_count):
    row = []
    for _ in range(row_count):
        row.append(list(input()))
    building.append(row)
    
for f in range(floor_count):
        for r in range(row_count):
            for c in range(col_count):
                if building[f][r][c] == 'S':
                    start =[f,r,c]
                    break

dr = [-1,0,1,0]
dc = [0,-1,0,1]
df = [-1,1]
depth =[]

for _ in range(floor_count):
    depth.append([[0]*col_count for _ in range(row_count)])
print(bfs(start))
```
