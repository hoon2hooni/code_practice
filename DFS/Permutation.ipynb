```python
ans = []
def dfs(elements,visited):
    global ans
    if len(elements) == 0:
        ans.append(visited[:])
    for e in elements:
        next_elements = elements[:]
        next_elements.remove(e)
        visited.append(e)
        dfs(next_elements,visited)
        visited.pop()
    

        
```

```python
dfs([1,2,3,4,5],[])
print(ans)
```
