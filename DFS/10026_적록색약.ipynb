```python
class DFS:
    
    def __init__(self, n ,matrix,sickness):
        self.matrix = copy.deepcopy(matrix)
        self.sickness =sickness
        self.length = n
        self.move = [(-1,0),(0,-1),(1,0),(0,1)]
        self.count = 0
        
    def search(self, start_x, start_y , color):
        pprint.pprint(self.matrix, width = 30)
        stack =[[start_x,start_y]]
        self.matrix[start_x][start_y] = 0
        while stack:
            print(stack, color)
            x, y = stack.pop()
            for i in range(4):
                next_x = self.move[i][0]
                next_y = self.move[i][1]
                if 0 <= next_x < self.length and \
                0 <= next_y < self.length:
                    if self.sickness == True:
                        if color == 'R' or color == 'G':
                            if self.matrix[next_x][next_y] in ['R','G']:
                                self.matrix[next_x][next_y] = 0
                                stack.append([next_x, next_y])                                
                        else:
                            if self.matrix[next_x][next_y] == color:
                                self.matrix[next_x][next_y] = 0
                                stack.append([next_x, next_y])  
                    else:
                        if self.matrix[next_x][next_y] == color:
                            self.matrix[next_x][next_y] = 0
                            stack.append([next_x, next_y]) 
        return 1
            
        
    def main(self):
        for i in range(self.length):
            for j in range(self.length):
                if self.matrix[i][j] == 'R' or self.matrix[i][j] =='G'\
                or self.matrix[i][j] == 'B':
                    self.count += self.search(i, j, self.matrix[i][j])
        return self.count
                
                
        
        
    
```
